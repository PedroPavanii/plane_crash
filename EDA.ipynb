### Drop columns that are not relevant:

```python
df = df.drop(['MSN', 'YOM', 'Time','Registration', 'Schedule', 'Flight no.', 'Crash location',
       'Crew fatalities',  'Other fatalities','Circumstances'], axis=1)
```

### Transforming date column into datetime and creating new year column:

```python
df['Year'] = df['Date'].apply(lambda x: x.split('-')[0])
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')
crash = df.drop(['Date'], axis = 1)
```

### Unifying values of the Flight type column:

```python
crash['Flight type'] = crash['Flight type'].str.replace('Scheduled Revenue Flight', 'Airline Flight')
crash['Flight type'] = crash['Flight type'].str.replace('Survey / Patrol / Reconnaissance', 'Military')
crash['Flight type'] = crash['Flight type'].str.replace('Bombing', 'Military')
crash['Flight type'] = crash['Flight type'].str.replace('Fire fighting', 'Military')
crash['Flight type'] = crash['Flight type'].str.replace('Survey / Patrol / Reconnaissance', 'Military')
crash['Flight type'] = crash['Flight type'].str.replace('Aerial photography', 'Military')
crash['Flight type'] = crash['Flight type'].str.replace('Delivery', 'Government')
crash['Flight type'] = crash['Flight type'].str.replace('Postal (mail)', 'Government')
crash['Flight type'] = crash['Flight type'].str.replace('Charter/Taxi (Non Scheduled Revenue Flight)', 'Private')
crash['Flight type'] = crash['Flight type'].str.replace('Executive/Corporate/Business', 'Private')
crash['Flight type'] = crash['Flight type'].replace('Postal (mail)', 'Government')
crash['Flight type'] = crash['Flight type'].replace('Charter/Taxi (Non Airline Flight)','Private')
```

### Unifying values of the Crash cause column:

```python
crash['Crash cause'] = crash['Crash cause'].replace('Unknown', 'Other causes')
```
